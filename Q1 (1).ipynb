```python
stats.chi2_contingency(chii)
```

As the expected frequencies are greater than 5 the chi - square results can be trusted.
We can reject the null hypothesis as the p-value is less than 0.05. Thus, the results indicate that there is a relationship between the presence of video evidence and the investigation to be completed

```python
video_table = pd.crosstab(index=dat['Complaint Has Video Evidence'], columns="count")
investigation_table = pd.crosstab(index=dat['Is Full Investigation'], columns="count")
print( "video")
print(video_table)
print(" ")
print( "completed")
print(investigation_table)
```

```python
observed = investigation_table

video_ratios = video_table/len(dat['Complaint Has Video Evidence'])  # Get population ratios

expected = video_ratios * len(dat['Is Full Investigation'])   # Get expected counts

chi_squared_stat = (((observed-expected)**2)/expected).sum()

print(chi_squared_stat)
```

```python
crit = stats.chi2.ppf(q = 0.95, # Find the critical value for 95% confidence*
                      df = 1)   # Df = number of variable categories - 1

print("Critical value")
print(crit)

p_value = 1 - stats.chi2.cdf(x=chi_squared_stat,  # Find the p-value
                             df=1)
print("P value")
print(p_value)
```
